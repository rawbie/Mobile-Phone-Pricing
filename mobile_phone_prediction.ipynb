# ***Treating the outliers***

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
columns_with_outliers = ['fc', 'px_height', 'three_g']

for col in columns_with_outliers:
    df=remove_outliers_iqr(data_frame,col)
```

# feature engineering

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['total_px'] = df['px_width'] * df['px_height']
```

```python
import numpy as np
df['screen_size'] = np.sqrt(df['sc_h']**2 + df['sc_w']**2)
```

```python
df['ppi'] = np.sqrt(df['px_height']**2 + df['px_width']**2) / df['screen_size']
```

```python
df['cam_score'] = df['pc']*0.7 + df['fc']*0
```

```python
df['battery_per_gram'] = df['battery_power'] / df['mobile_wt']
```

```python
df.drop(columns=['px_height', 'px_width', 'sc_h', 'sc_w', 'pc', 'fc'], inplace=True)
```

```python
df.shape
```

# ***Model Training and Splitting***

```python
X = df.drop(columns=['price_range'],axis=1)
y = df['price_range']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

## ***Logistic Model***

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
logistic_model=LogisticRegression(multi_class='ovr')
```

```python
logistic_model.fit(x_train,y_train)
```

```python
y_pred_log=logistic_model.predict(x_test)
```

```python
print(" Logistic Regression\n")
print("-"*75,"Accuracy","-"*75)
print(" Accuracy:", accuracy_score(y_test, y_pred_log))

print("-"*75,"Classification Report","-"*75)
print(" Classification Report:\n", classification_report(y_test, y_pred_log))
print("-"*75,"Confusion Matrix","-"*75)
cm_log = confusion_matrix(y_test, y_pred_log)
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## ***DECESION TREE CLASSIFIER***

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
import seaborn as sns
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best'],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2'],
    'random_state': [42]
}
```

```python
decision_tree_model=DecisionTreeClassifier(random_state=42)
```

```python
grid = GridSearchCV(decision_tree_model, params, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
```

```python
grid.fit(x_train,y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
y_pred_decision_tree=grid.predict(x_test)
```

```python
print(" Decision Tree Model\n")
print("-"*75,"Accuracy","-"*75)
print(" Accuracy:", accuracy_score(y_test, y_pred_decision_tree))
print("-"*75,"Classification Report","-"*75)
print(" Classification Report:\n", classification_report(y_test, y_pred_decision_tree))
print("-"*75,"Confusion Matrix","-"*75)
cm_log = confusion_matrix(y_test, y_pred_decision_tree)
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
